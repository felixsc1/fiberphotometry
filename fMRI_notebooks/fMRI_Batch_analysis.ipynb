```python
# splitall function from: https://www.safaribooksonline.com/library/view/python-cookbook/0596001673/ch04s16.html
def splitall(path):
    allparts = []
    while 1:
        parts = os.path.split(path)
        if parts[0] == path:  # sentinel for absolute paths
            allparts.insert(0, parts[0])
            break
        elif parts[1] == path: # sentinel for relative paths
            allparts.insert(0, parts[1])
            break
        else:
            path = parts[0]
            allparts.insert(0, parts[1])
    return allparts


for folderName, subfolders, filenames in os.walk(file_path):
    for file in filenames:
        if (fnmatch.fnmatch(file, '*+orig.BRIK*') and os.path.getsize(os.path.join(folderName,file))>1000000
           and 'analyze' not in folderName):
            ID = splitall(folderName)[4]+'_scan'+splitall(folderName)[5]
            file_path1 = glob.glob(os.path.join(folderName,file))
            nb_kwargs = {'file_path': file_path1[0], 'ID': ID}
            print('running analysis for...', file_path1[0])
            print(ID)
            run_notebook(nb_name, save_ipynb=False, nb_kwargs=nb_kwargs)
            
            # then copy the stats and ROI files in one result folder
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','*stats*BRIK')):
                shutil.copy(resultfiles, groupstatsfolder)
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','*stats*HEAD')):
                shutil.copy(resultfiles, groupstatsfolder)
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','ROIdata*')):

                shutil.copy(resultfiles, groupstatsfolder)
```

```python
# Fixing baseline - removing linear trend 
from scipy import signal

def detrend_baseline(array,bsl_dur):
    xx=signal.detrend(array[0:bsl_dur])
    coeffs = np.polyfit(np.arange(0,bsl_dur), array[0:bsl_dur], 1)
    array[0:bsl_dur]=xx
    array[bsl_dur:]+=2*coeffs[1]
    return array


# create new hierachical df for plotting:
forplotting = []
axis_time = np.arange(-onset,contra.shape[0]-onset)
forplotting_index = pd.Index(axis_time, name='Time [s]')
columns = pd.MultiIndex.from_product([['ipsi', 'contra'],['mean','SEM']],names=['ROI','signal'])
forplotting = pd.DataFrame(np.zeros((forplotting_index.size, columns.size)), index=forplotting_index, columns = columns)

# insert detrended mean values
forplotting.loc[:,('contra','mean')] = detrend_baseline(contra_all.mean(axis=1).values,600)   #not full baseline just 10min in case it creates some dip
forplotting.loc[:,('ipsi','mean')] = detrend_baseline(ipsi_all.mean(axis=1).values,600)

# and Standart Error (SEM) + error bands
forplotting.loc[:,('contra','SEM')] = contra_all.sem(axis=1).values
forplotting.loc[:,('ipsi','SEM')] = ipsi_all.sem(axis=1).values
forplotting['contra','upper'] = forplotting['contra','mean']+forplotting['contra','SEM']
forplotting['contra','lower'] = forplotting['contra','mean']-forplotting['contra','SEM']
forplotting['ipsi','upper'] = forplotting['ipsi','mean']+forplotting['ipsi','SEM']
forplotting['ipsi','lower'] = forplotting['ipsi','mean']-forplotting['ipsi','SEM']  

```
