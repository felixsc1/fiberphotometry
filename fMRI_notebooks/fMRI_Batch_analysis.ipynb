```python
import glob
import os, sys, shutil
from nbrun import run_notebook #downloaded from https://github.com/tritemio/nbrun/blob/master/example-master-notebook.ipynb
from tkinter import Tk
from tkinter.filedialog import askdirectory
import fnmatch
```

```python
nb_name = 'fMRI_Analysis_2018.ipynb'
```

```python
root = Tk()
ttl = "Select main directory with all animals"

dir1 = '/home/aic/BigHDD/MR_DATA/CSI2018'
    
root.withdraw()
file_path = askdirectory(initialdir = dir1, title = ttl)
root.update()
root.destroy()
print('main path :', file_path)

# os.mkdir(groupstatsfolder)
```

```python
# splitall function from: https://www.safaribooksonline.com/library/view/python-cookbook/0596001673/ch04s16.html
def splitall(path):
    allparts = []
    while 1:
        parts = os.path.split(path)
        if parts[0] == path:  # sentinel for absolute paths
            allparts.insert(0, parts[0])
            break
        elif parts[1] == path: # sentinel for relative paths
            allparts.insert(0, parts[1])
            break
        else:
            path = parts[0]
            allparts.insert(0, parts[1])
    return allparts


for folderName, subfolders, filenames in os.walk(file_path):
    for file in filenames:
        if (fnmatch.fnmatch(file, '*+orig.BRIK*') and os.path.getsize(os.path.join(folderName,file))>1000000
           and 'analyze' not in folderName):
            ID = splitall(folderName)[4]+'_scan'+splitall(folderName)[5]
            file_path1 = glob.glob(os.path.join(folderName,file))
            nb_kwargs = {'file_path': file_path1[0], 'ID': ID}
            print('running analysis for...', file_path1[0])
            print(ID)
            run_notebook(nb_name, save_ipynb=False, nb_kwargs=nb_kwargs)
            
            # then copy the stats and ROI files in one result folder
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','*stats*BRIK')):
                shutil.copy(resultfiles, groupstatsfolder)
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','*stats*HEAD')):
                shutil.copy(resultfiles, groupstatsfolder)
            for resultfiles in glob.glob(os.path.join(folderName,'analyzed','ROIdata*')):

                shutil.copy(resultfiles, groupstatsfolder)
```

```python
groupstatsfolder = '/home/aic/BigHDD/MR_DATA/CSI2018/groupresults'

file_ending = "*.1D"
ROIfiles = glob.glob(os.path.join(os.path.abspath(groupstatsfolder),file_ending))

dataframes = []
keys = []
for files in ROIfiles:
    dataframes.append(pd.read_table(files))
    keys.append(dataframes[-1]['File'][0][11:-10])
    del dataframes[-1]['File']
    del dataframes[-1]['Sub-brick']

frames = pd.concat(dataframes, axis=1, keys=keys)
```

```python
filteredframes=frames.filter(regex='1mA', axis=1)

contra=filteredframes.iloc[:, filteredframes.columns.get_level_values(1)=='Mean_S1hlContra  ']
ipsi=filteredframes.iloc[:, filteredframes.columns.get_level_values(1)=='Mean_S1hlIpsi  ']

filteredframes.to_excel(os.path.join(os.path.abspath(groupstatsfolder),'All_ROItraces.xlsx'))


forplotting_contra=contra.mean(axis=1)
forplotting_ipsi=ipsi.mean(axis=1)
```
